1. How the AI "Learns" (The Logic)
Instead of teaching the AI from scratch, I used Transfer Learning.

I took a pre-trained model called MobileNetV2 which was already trained on millions of images.

Even though I only had 70 images per animal, the AI was successful because it already knew how to recognize shapes, textures, and edges from its previous training. I just had to teach it how to connect those shapes to these specific 5 animals.

2. Handling Small Data (Augmentation)
To make the most of my 70 images, I used Data Augmentation.

The code automatically flipped, rotated, and zoomed into my photos during training.

This "tricked" the AI into seeing thousands of different variations instead of just the original 70, which prevented the model from simply memorizing the photos (Overfitting).

3. Folder Structure & Pipeline
The project follows a standard machine learning pipeline:

Training Set: Used to teach the model.

Validation Set: Used as a "practice exam" to check accuracy during training.

Test Set: Used for final verification.

4. Final Result
The model achieved a Validation Accuracy of 98.8%, and when tested with a random image, it correctly identified a Cat with 100% confidence.

```
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

# 1. PREPARE THE DATA
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Tell the code to look inside 'Animal_Data/train'
train_generator = train_datagen.flow_from_directory(
    'Animal_Data/train',
    target_size=(224, 224),
    batch_size=10,
    class_mode='categorical'
)

# Tell the code to look inside 'Animal_Data/validation'
val_generator = val_datagen.flow_from_directory(
    'Animal_Data/validation',
    target_size=(224, 224),
    batch_size=10,
    class_mode='categorical'
)

# 2. BUILD THE MODEL Using Transfer Learning
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False # Don't change the base "knowledge"

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dense(5, activation='softmax') # '5' because there are 5 animals
])

# 3. TRAIN THE AI
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("Training starting...")
model.fit(train_generator, validation_data=val_generator, epochs=10)
```

```
import numpy as np
from google.colab import files
from tensorflow.keras.preprocessing import image

# 1. Get the names of your animals in order
class_names = sorted(list(train_generator.class_indices.keys()))

# 2. Upload a photo from your computer
uploaded = files.upload()

for fn in uploaded.keys():
  # 3. Prepare the image for the AI
  path = fn
  img = image.load_img(path, target_size=(224, 224))
  x = image.img_to_array(img)
  x = np.expand_dims(x, axis=0)
  x = x / 255.0  # Remember to normalize!

  # 4. Make the prediction
  predictions = model.predict(x)
  score = predictions[0]

  # 5. Show the result
  print(f"\nI am {100 * np.max(score):.2f}% sure this is a {class_names[np.argmax(score)]}")
```
